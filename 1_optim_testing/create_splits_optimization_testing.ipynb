# EEGMAT cross_json optimization benchmark

This notebook stages EEGMAT data to local disk, runs baseline vs optimized split creation, times each 3x, and compares outputs.

```python
from google.colab import drive
drive.mount("/content/drive", force_remount=False)
```

```python
import os
import subprocess
import sys

repo_root = "/content/AdaBrain-Bench"
if not os.path.exists(repo_root):
    subprocess.run(["git", "clone", "--depth", "1", "https://github.com/iroblesrazzaq/AdaBrain-Bench.git", repo_root], check=True)
else:
    subprocess.run(["git", "-C", repo_root, "pull"], check=True)
subprocess.run(["git", "-C", repo_root, "log", "-1"], check=True)
```

```python
data_root_drive = "/content/drive/MyDrive/AdaBrain-Bench Data"
data_root_local = "/content/AdaBrain-Bench-Data-local"
src = os.path.join(data_root_drive, "EEGMAT", "processed_data")
dst = os.path.join(data_root_local, "EEGMAT", "processed_data")
if not os.path.exists(src):
    raise FileNotFoundError(f"Missing source data: {src}")
os.makedirs(dst, exist_ok=True)
subprocess.run(["bash", "-lc", f"rsync -av --delete \"{src}/\" \"{dst}/\""], check=True)
```

```python
benchmark_script = os.path.join(repo_root, "1_optim_testing", "benchmark_eegmat_cross_json.py")
subprocess.run([
    sys.executable, benchmark_script,
    "--data-root", data_root_local,
    "--repo-root", repo_root,
    "--runs", "3"
], check=True)
```
